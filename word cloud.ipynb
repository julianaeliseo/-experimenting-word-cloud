```python
import pandas as pd

df = pd.read_csv("./articles.csv") 
```

```python
df.head()
```

```python
import collections
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from matplotlib import rcParams
from wordcloud import WordCloud, STOPWORDS
%matplotlib inline
df['title']
```

```python
#stopwords are words to be desconsidered from wordcloud
stopwords = STOPWORDS
lista = ['por','para','em','diz','que','o','e','a','é','sobre','de']
for value in lista:
    stopwords.add(value)
```

```python
headlines = ' '.join(df['title'].str.lower())
wordcloud = WordCloud(stopwords=stopwords, background_color="black", max_words=1000).generate(headlines)
```

```python
rcParams['figure.figsize'] = 10, 20
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
